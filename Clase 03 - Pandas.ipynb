#### **Pandas** es una librería de Python que provee herramientas para análisis de datos, proporciona estructuras de datos flexibles y muy eficientes.

 - Series: Arreglos unidimensionales (P. ej: Series temporales)
 - DataFrame: Son estructuras de datos similares a las tablas.
 - Panel, Panel4D y PanelND: Para trabajar con datos multidimensiones (> 2 dim).

### Importamos la librería

Es convención utilizar el alias **pd** cuando se va a trabajar con **pandas**

```python
import pandas as pd
```

#### Añadir columnas desde otro DataFrame compartiendo una columna (generalmente un Id)     `join()` o `merge()`

```python
df_1 = df.iloc[:, :3] #Creamos un subconjunto de datos que tenga algunas columnas (incluyendo la columna "Id")
df_1
```

```python
df_2 = df.iloc[:, 3:] # Creamos otro subconjunto que contenga las columnas restantes
df_2 = df_2.assign(Id=df[["Id"]]) #Agregamos la columna "Id" al final para que los subconjuntos tengan una columna en común
df_2
```

```python
df_3 = df_1.join(df_2, on=["Id"], rsuffix="rId") #Realizamos el join sobre la columna "Id", manteniendo la columna Id de la 2da tabla con el nombre "rId"
```

```python
df_3
```

Equivalente a la anterior (no duplica la columna en común)

```python
df_4 = df_1.merge(df_2, on="Id")
```

```python
df_4
```

#### Como crear un DataFrame

A través de un diccionario

```python
alturas = np.random.uniform(1.5, 2.0, 10)
edades = np.random.uniform(18, 99, 10)
datos = {"Altura": alturas, "Edad": edades} 
```

```python
df_5 = pd.DataFrame(data=datos) #Las columnas deben tener igual cantidad de datos
df_5
```

Creando un DataFrame vacío con los nombres de las columnas y completando los datos después

```python
df_6 = pd.DataFrame(columns=("Altura", "Edad"))
df_6
```

```python
df_6.Edad = edades
df_6
```

```python
df_6.Altura = alturas # Las columnas deben tener la misma cantidad de valores
df_6
```

### Graficar valores

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

Líneas

```python
df_6[["Edad"]].plot(kind='line')
```

Agregamos labels y cambiamos el color de la linea

```python
df_6[["Altura"]].plot(kind='line', color="orange")
plt.ylabel("Altura en mts")
plt.xlabel("Individuo")
```

Scatter - Cambiamos el color de los puntos y agregamos el título del gráfico

```python
df_6.plot(kind='scatter', x="Edad", y="Altura", color="green", title="Alturas y Edades")
```

Barras - desactivamos la leyenda del gráfico

```python
df_6.plot(kind='bar', x="Edad", y="Altura", legend=False)
```

Guardamos un plot con `plt.savefig()`

```python
#rcParams.figSize = (10, 20)
```

```python
ax = df_6.plot(kind='bar', x="Edad", y="Altura", legend=False, title="Altura y edades")
ax.set_ylabel("Altura")
ax.set_xlabel("Edad")
plt.savefig("mi_grafico_de_barras.png", dpi=300, bbox_inches='tight')
```
