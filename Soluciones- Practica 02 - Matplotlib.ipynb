```python
import matplotlib.pyplot as plt
import numpy as np
```

## Graficar la siguiente función:

```python
X = np.arange(1, 101)
Y =  np.random.random_sample((100))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)# *nrows*, *ncols*, and *index*
ax.set_title('UN TITULO')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(X, Y, ".r")
ax.grid()
```

## Graficar la siguiente función:

```python
X = np.arange(1, 101)
Y =  np.random.lognormal(size=(100))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)# *nrows*, *ncols*, and *index*
ax.set_title('UN TITULO')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(X, Y, ".r")
ax.grid()
```

## Dado $F(x) = ax + b$ con $x > 1$ y $x<=100$, graficar:

### Grafico 1:  $a=100$ y $b=0.5$
### Grafico 2:  $a=0.5$ y $b=100$

```python
X = np.arange(1,101)
```

```python
data_gf_1 =  (100*X)+ 0.5
```

```python
data_gf_2 =  (0.5*X) + 100
```

```python
fig = plt.figure()
ax = fig.add_subplot(121)# *nrows*, *ncols*, and *index*
ax.set_title('GRAFICO 1')
ax.plot(X, data_gf_1)
ax.set_ylabel('Y')
ax.set_xlabel('X')

ax_2 = fig.add_subplot(122)# *nrows*, *ncols*, and *index*
ax_2.set_title('GRAFICO 2')
ax_2.set_xlabel('X')
ax_2.plot(X, data_gf_2, "r-")
```

## Realizar un grafico de barras:

```python
data = {
    'Nombre':['Juan', 'Caro', 'Rodrigo', 'Ana', 'Virginia'],
    'Altura':[2, 1.5, 1.7, 1.66, 1.90]
}
```

```python
data["Nombre"]
```

```python
data['Altura']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
x_value = [0, 1, 2, 3, 4]
y_value = data["Altura"]
ax.bar(x_value, y_value, align='center')
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(data["Nombre"])
ax.set_ylabel('Altura (metros)')
ax.set_title('Alturas')
```

## Descargar una imagen de internet y mostrarla

```python
img = plt.imread("logopythonred.jpg")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(img)
```

```python
img.shape
```

## Realizar un corte a la imagen 

```python
crop =  img[0:512,512:1024]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(crop)
```

## Rotar una imagen 

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_yticks([])
ax.set_xticks([])
ax.imshow(np.flip(crop))
```
